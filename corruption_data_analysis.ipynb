## Data Analysis of Jakarta Corruption Cases

```python
import numpy as np
import pandas as pd
import pickle
```

```python
# Open case list
directory = 'C:/Users/akmal.fadhlurrahman/Documents/personal_stuff/adhoc_judge/data/'
case = pickle.load(open(directory+'case_list.pickle','rb'))
judge = pickle.load(open(directory+'case_judge_list.pickle','rb'))
clerk = pickle.load(open(directory+'case_clerk_list.pickle','rb'))
defendants = pickle.load(open(directory+'terdakwa_table.pickle','rb'))
prosecutors = pickle.load(open(directory+'jaksa_table.pickle','rb'))
info = pickle.load(open(directory+'general_info_table.pickle','rb'))
sentence = pickle.load(open(directory+'sentence_table.pickle','rb'))
```

```python
case.tail()
```

```python
# JUDGE
judge = judge[judge['Aktif'] == 'Ya'].reset_index()
judge['No']  = judge.sort_values(['Nomor Perkara','Posisi','Aktif'],ascending=[True,False,True]) \
                    .groupby(['Nomor Perkara','Posisi']) \
                    .cumcount() + 1
judge['Nomor Posisi'] = judge['Posisi'] + '_' + judge['No'].map(str)
judge = judge.pivot(index='Nomor Perkara', columns='Nomor Posisi', values='Nama Hakim/Majelis Hakim')
judge.head()
```

```python
# Clerk
clerk = clerk[clerk['Aktif'] == 'Ya'].reset_index()
clerk['No']  = clerk.sort_values(['Nomor Perkara','Aktif'],ascending=[True,True]) \
                    .groupby(['Nomor Perkara']) \
                    .cumcount() + 1

clerk['Nomor Posisi'] = 'Panitera_' + clerk['No'].map(str)
clerk = clerk.pivot(index='Nomor Perkara', columns='Nomor Posisi', values='Nama Panitera Pengganti')
clerk.head()
```

```python
# prosecutors
prosecutors['Nomor Posisi'] = 'Jaksa_' + prosecutors['No'].map(str)
prosecutors = prosecutors.reset_index().pivot(index='Nomor Perkara', columns='Nomor Posisi', values='Nama')
prosecutors.head()
```

# Joining Tables

```python
# Rename defendant name on table
defendants = defendants.rename(index=str,columns={'Nama':'Nama Terdakwa'})
```

```python
# Start from defendant table and left join to case
result = defendants.merge(case[['Tanggal Register','Status Perkara','Lama Proses']],
                          left_index = True,
                          right_index = True,
                          how='left')
```

```python
# Then left join to info table to get details of indictment
result = result.merge(info[['Tanggal Pendaftaran','Dakwaan']],
                     left_index = True,
                     right_index = True,
                     how = 'left')
result[['Dakwaan']] = result[['Dakwaan']].replace(r'\\n',' ',regex=True)
```

```python
# left join to sentence table
result = result.merge(sentence[['Amar Putusan','Tanggal Minutasi','Tanggal Putusan']],
                     left_index = True,
                     right_index = True,
                     how = 'left')
```

```python
# Left join with judge table
result = result.merge(judge,
                     left_index = True,
                     right_index = True,
                     how = 'left')
```

```python
# Left join with prosecutors
result = result.merge(prosecutors,
                     left_index = True,
                     right_index = True,
                     how = 'left')
```

```python
# Left join with clerk
result = result.merge(clerk,
                     left_index = True,
                     right_index = True,
                     how = 'left')
```
